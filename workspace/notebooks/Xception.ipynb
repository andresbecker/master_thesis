```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pretrained_model = tf.keras.applications.Xception(
    include_top=False,
    weights="imagenet",
    input_tensor=None,
    input_shape=None,
    pooling=None,
    classifier_activation=None,
)
```

```python
pretrained_model.layers[4].get_weights()[0].shape
```

```python
layer = pretrained_model.layers[1]
fist_conv_w = layer.get_weights()[0]
print(fist_conv_w.shape)
a = np.min(fist_conv_w)
b = np.max(fist_conv_w)
fist_conv_w_norm = (fist_conv_w - a) / (b - a)
n_filters = fist_conv_w_norm.shape[-1]
print(n_filters)

plt.figure(figsize=(8*4,8*4))
for i in range(n_filters):
    plt.subplot(8,8,i+1)
    img = fist_conv_w_norm[:,:,:,i]
    plt.imshow(img,
               cmap=None,
               vmin=0, vmax=1)

plt.show()
```
